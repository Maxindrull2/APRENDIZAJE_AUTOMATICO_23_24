# **Metodología CRISP-DM**

CRISP-DM (Cross-Industry Standard Process for Data Mining), es una metodología estructurada y flexible, que permite adaptarse a una variedad de industrias y casos de uso.

Una de las principales razones por las que se utiliza CRISP-DM es su capacidad para manejar datos de manera iterativa y cíclica, lo cual permite volver hacia atrás a fases anteriores si es necesario. Esto contrasta con otras metodologías, como KDD (Knowledge Discovery in Databases), que tienden a ser más lineales. La ventaja de CRISP-DM radica en su enfoque práctico y orientado a los negocios, lo que facilita la integración de los resultados de la minería de datos en las decisiones de negocio.

Una de las principales diferencias entre CRISP-DM y KDD es que CRISP-DM incorpora una fase específica de comprensión del negocio y una fase de despliegue. Esto asegura que el proyecto esté alineado con los objetivos desde el principio y que los resultados sean aplicables y útiles en su contexto.

Este conjunto de datos se basa en la transcripción de informes originales de crímenes, los cuales fueron originalmente escritos a máquina. Para un análisis efectivo de la base de datos, se procede a comprender los atributos que contiene. A continuación se presenta una explicación detallada de cada columna de la misma:

*   **DR_NO**: Este es el número oficial del expediente que consta de un año de dos dígitos, un ID de área y cinco dígitos adicionales. Representa el "Division of Records Number", importante para la identificación única de cada caso registrado.

*   **Date Rptd**: La fecha en que se reportó el crimen, en formato MM/DD/YYYY,  indicando cuándo se conocieron los hechos ante la autoridad.

*   **DATE OCC:** La fecha exacta en que ocurrió el crimen, también en formato MM/DD/YYYY. Es fundamental para determinar cuándo sucedieron específicamente los incidentes.

*   **TIME OCC:** La hora del crimen en formato militar de 24 horas.

*   **AREA:** Representa una de las 21 Estaciones de Policía Comunitaria de LAPD, numeradas secuencialmente del 1 al 21. Cada área geográfica tiene un número asignado importante para el análisis regional.

*   **AREA NAME:** Las 21 Áreas Geográficas o Divisiones de Patrulla también reciben un nombre que hace referencia a un punto de referencia o a la comunidad circundante a la que sirven. Por ejemplo, la División de la Calle 77 se encuentra en la intersección de South Broadway y la Calle 77, sirviendo a los barrios del sur de Los Ángeles.

*   **Rpt Dist No:** Número de cuatro dígitos que representa una subárea dentro de una Área Geográfica.

*   **Part 1-2:** Su significado no está especificado en la fuente original de la base de datos, sin embargo, según fuentes externas este atributo puede indicar si el crimen es de Parte 1 (más grave) o Parte 2 (menos grave), útil para clasificar la severidad del delito.

*   **Crm Cd:** Código numérico que indica el crimen cometido, importante para la categorización de los delitos.

*   **Crm Cd Desc:** Descripción textual del código de crimen, proporciona claridad sobre la naturaleza del delito reportado.

*   **Mocodes:** Mocodes: Modus Operandi: Actividades asociadas con el sospechoso en la comisión del crimen. Por ejemplo, el código 0102, 'blind', podría indicar que el sospechoso intenta cegar a la víctima o a testigos para evitar ser identificado.

  Cada Mocode se asigna a diferentes acciones específicas que el sospechoso podría haber utilizado durante la comisión del delito, desde métodos de engaño hasta el uso de fuerza física o armas.

  Además, el análisis de los Mocodes puede revelar tendencias en la actividad criminal y ajustes en las tácticas delictivas, lo que permite a las fuerzas de seguridad adaptar sus métodos de patrullaje y sus enfoques investigativos. Por ejemplo, un aumento en el uso de ciertos Mocodes puede indicar la emergencia de nuevas formas de criminalidad o un cambio en las estrategias de los delincuentes

*   **Vict Age:** Edad de la víctima, representada por dos caracteres numéricos.

*   **Vict Sex:** Sexo de la víctima, identificado como F (Femenino), M (Masculino), X (Desconocido).

*   **Vict Descent:** Código de ascendencia de la víctima, con múltiples códigos representando diferentes grupos étnicos y raciales.
Algunos de estos códigos son: A para Otra raza de Asiático, B para Negro, C para Chino, D para Camboyano, F para Filipino, G para Guameño, H para Hispano/Latino/Mexicano, I para Indio Americano o Nativo de Alaska, y J para Japonés.

*   **Premis Cd**: Código numérico que identifica el tipo de estructura, vehículo o ubicación donde se cometió el crimen.

*   **Premis Desc:** Descripción del Código de Premisas, detallando más sobre el lugar del incidente.

*   **Weapon Used Cd:** Código del tipo de arma usada en el crimen, si aplica.

*   **Weapon Desc:** Descripción del arma utilizada, proporciona detalles adicionales sobre el arma reportada.

*   **Status**: Estado actual del caso, por ejemplo, "IC" como predeterminado que indica "Investigación Continua".

*   **Status Desc**: Descripción del estado del caso, aporta más detalles sobre la etapa actual de la investigación.

*   **Crm Cd 1, Crm Cd 2, Crm Cd 3, Crm Cd 4:** Códigos adicionales que pueden contener crímenes menos graves relacionados con el incidente principal. Son útiles para entender si hubo múltiples infracciones durante un solo evento.

*   **LOCATION:** Dirección aproximada del incidente del crimen, redondeada al bloque más cercano para mantener el anonimato.

*   **Cross Street:** Calle transversal más cercana a la dirección redondeada, útil para localizar el área de interés.

*   **LAT y LON**: Coordenadas de latitud y longitud del lugar del crimen, esenciales para el mapeo geográfico y análisis espacial de la actividad criminal.

Estos atributos permiten una comprensión detallada y profunda de cada incidente reportado, facilitando análisis complejos sobre patrones de crimen. Además, ofrece información valiosa para observar los factores temporales, espaciales y propios de las víctimas.

### Fecha

A partir de la fecha se determian el día de la semana y el mes en el que ocurrio el delito

```python
def obtener_dia_semana(fecha_str):
    fecha_dt = datetime.strptime(fecha_str, '%m/%d/%Y %I:%M:%S %p')
    dia_semana = fecha_dt.weekday()
    dias_semana = ["lunes", "martes", "miércoles", "jueves", "viernes", "sábado", "domingo"]
    return dias_semana[dia_semana]

def obtener_mes(fecha_str):
    fecha_dt = datetime.strptime(fecha_str, '%m/%d/%Y %I:%M:%S %p')
    return fecha_dt.strftime("%B")  # Devuelve el nombre completo del mes

df['Dia_Semana'] = df['DATE OCC'].apply(obtener_dia_semana)
df['Month'] = df['DATE OCC'].apply(obtener_mes)
```

## **Regresión Logística:**

La **regresión logística** es una técnica estadística utilizada para modelar la relación entre una variable dependiente categórica y una o más variables independientes. Es particularmente adecuada cuando la variable de respuesta Y es categórica múltiple, es decir, cuando puede tomar varios valores discretos. Sin embargo, es especialmente útil cuando la variable de respuesta solo tiene dos posibles resultados (binaria).


La regresión logística se basa en la **función logística**, que se expresa de la siguiente manera:

$P(C = 1 \mid \mathbf{x}) = \frac{1}{1 + e^{-(\beta_0 + \sum_{i=1}^{n} \beta_i x_i)}}$

donde P(C = 1 | x) es la probabilidad de que el evento C ocurra dado un conjunto de variables independientes x. Los parámetros beta_0, beta_1,..., beta_n se estiman a partir de los datos y son los coeficientes de la regresión.

La función logística transforma una combinación lineal de los predictores a un rango entre 0 y 1, produciendo una curva sigmoidal que se utiliza para modelar probabilidades.


#### **Sesgo en la Regresión Logística**

El **sesgo** en la regresión logística se refiere a la desviación sistemática de las estimaciones del modelo con respecto a los valores reales. Este sesgo puede surgir de diversas fuentes. Una de las principales causas es la **no representatividad de los datos** de entrenamiento respecto a la población general, lo que puede resultar en predicciones sesgadas si ciertos grupos están subrepresentados o sobrerepresentados. Además, la **omisión de variables relevantes**, que afectan a la variable respuesta, puede inducir sesgo, ya que el modelo no captura todas las influencias significativas sobre la variable dependiente. La **multicolinealidad**, cuando las variables independientes están altamente correlacionadas entre sí, también puede introducir sesgo al provocar inestabilidad en las estimaciones de los coeficientes. Por último, la **selección de muestras no aleatoria**, donde los datos utilizados para entrenar el modelo no son una muestra aleatoria de la población, puede llevar a un sesgo de selección, afectando la generalizabilidad del modelo a nuevas muestras. Es crucial abordar estos problemas mediante la cuidadosa selección y preprocesamiento de datos, así como mediante la validación adecuada del modelo para asegurar su robustez y precisión.

#### **Ventajas y Limitaciones**

La regresión logística, usada para predecir una variable categórica binaria basada en varias variables predictoras, ofrece varias ventajas, como su interpretación intuitiva, donde los coeficientes estimados pueden ser entendidos como los efectos log-odds de los predictores en la probabilidad del evento, y su capacidad para proporcionar probabilidades directas, lo que facilita la toma de decisiones. Además, es flexible, ya que puede manejar tanto variables continuas como categóricas. Por lo tanto, esta técnica permite interpretar los coeficientes de regresión para entender la influencia de cada variable independiente en la probabilidad del evento, así como predecir resultados binarios en situaciones donde la relación entre la variable dependiente y las independientes no es lineal.

Sin embargo, presenta algunas limitaciones, incluyendo la suposición de una relación lineal entre las variables independientes y el logit de la variable dependiente. La alta correlación entre variables independientes (multicolinealidad) puede dificultar la estimación precisa de los coeficientes, y cuando las clases están desbalanceadas, el modelo puede tener dificultades para predecir correctamente la clase minoritaria.

### **Teoría de K-means**

K-means es una técnica de clustering no supervisado que agrupa datos en un número predefinido de clústeres, basándose en la similitud entre los elementos. Este método es ampliamente utilizado para identificar grupos naturales dentro de grandes conjuntos de datos en diversas aplicaciones, desde la segmentación de mercado hasta la organización de bases de datos.

Funcionamiento del Algoritmo

El algoritmo K-means organiza los datos en clústeres basándose en la proximidad a los centroides de estos clústeres, que son puntos representativos calculados como el promedio de los puntos en cada clúster. A lo largo del proceso, K-means busca minimizar la suma de las distancias cuadradas entre cada punto y el centroide de su clúster asignado.

Pasos del Algoritmo:

*   Inicialización: Comienza seleccionando aleatoriamente K puntos del conjunto de datos como centroides iniciales. Alternativamente, técnicas como K-means++ pueden ayudar a mejorar la selección inicial para evitar mínimos locales no óptimos.

*  Asignación: Cada punto del conjunto de datos es asignado al clúster cuyo centroide es el más cercano, usualmente usando la distancia euclidiana como métrica de similitud.

*  Actualización de centroides: Una vez todos los puntos han sido asignados a un clúster, se recalcula el centroide de cada clúster tomando el promedio de todos los puntos que han sido asignados al clúster.

*  Iteración: Los pasos de asignación y actualización se repiten hasta que los centroides ya no cambian significativamente entre iteraciones, indicando que el algoritmo ha alcanzado la convergencia.

A continuación, se resaltan una serie de consideraciones que es importante tener en cuenta sobre el algoritmo.

*  Elección de K: Determinar el número adecuado de clústeres, K, es crucial y puede influir significativamente en los resultados. Métodos como el análisis del codo se utilizan frecuentemente para estimar el número óptimo de clústeres.

*  Sensibilidad a la inicialización: Los resultados de K-means pueden variar según la selección inicial de centroides. Realizar varias ejecuciones del algoritmo con diferentes inicializaciones puede ayudar a encontrar una solución más robusta y representativa.

*  Limitaciones: K-means asume que los clústeres son isotrópicos y convexos, lo cual puede no ser adecuado para algunos conjuntos de datos donde los clústeres reales tienen formas irregulares o tamaños muy variados.

```python
def find_closest_cluster_index(row, clusters):
    distances = cdist([row.tolist()], clusters, metric='euclidean')
    return np.argmin(distances)

X_test['Predicción'] = predictions_threshold
X_test['Real'] = y_test

df_cluster = X_test.copy()
df_cluster = df_cluster.drop('Real', axis = 1)

scaler = StandardScaler()
df_cluster[['Vict Age', 'Vict Age : F']] = scaled_numeric_features = scaler.fit_transform(df_cluster[['Vict Age', 'Vict Age : F']])

df_cluster_1 = df_cluster[df_cluster['Predicción'] == 1]
df_cluster_0 = df_cluster[df_cluster['Predicción'] == 0]

closest_clusters_1 = df_cluster_1.drop('Predicción', axis=1).apply(lambda row: find_closest_cluster_index(row, C_1), axis=1)
closest_clusters_0 = df_cluster_0.drop('Predicción', axis=1).apply(lambda row: find_closest_cluster_index(row, C_0), axis=1)

df_cluster.loc[df_cluster['Predicción'] == 1, 'Predicción'] = np.where(closest_clusters_1.isin([0,6,7,8]), -1, 1) # [0,6,7,8]
df_cluster.loc[df_cluster['Predicción'] == 0, 'Predicción'] = np.where(closest_clusters_0.isin([4,6,8,17]), 1, 0) # [4,6,8,17]
df_cluster.loc[df_cluster['Predicción'] == -1, 'Predicción'] = 0
```

Por tanto, es fundamental aumentar la vigilancia y la presencia policial durante las horas nocturnas, especialmente en áreas de alto riesgo. Campañas de concienciación sobre los riesgos de salir solos durante la noche y fomentar el uso de servicios de transporte seguros pueden ayudar a mitigar este riesgo. La colaboración con establecimientos nocturnos para mejorar las medidas de seguridad y asistencia también puede ser efectiva.
